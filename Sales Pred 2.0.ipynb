```python
month_year = []
for i in df['order_date']:
    a = i.strftime('%m-%Y')
    month_year.append(a)
df['month_year'] = month_year 
```

```python
trend = df.groupby("month_year").sum()['sales'].reset_index()
```

```python
plt.figure(figsize=(18,7))
plt.plot(trend['month_year'],trend['sales'],color="blue")
plt.xticks(rotation='vertical')
plt.xlabel("Month-Year")
plt.ylabel("Sales")
plt.show()
```
